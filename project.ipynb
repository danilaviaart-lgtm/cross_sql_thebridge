```python
# import sqlite

import sqlite3

import pandas as pd
```

```python
con = sqlite3.connect("src/thebridge.db")
```

```python
cur = con.cursor()
```

```python
cur.execute("""
CREATE TABLE campus (
    id_campus INT PRIMARY KEY,
    nombre_campus VARCHAR(25) NOT NULL
);
""")
```

```python
cur.execute("""
CREATE TABLE vertical (
    id_vertical INT PRIMARY KEY,
    nombre_vertical VARCHAR(25) NOT NULL
);
""")
```

```python
cur.execute("""
CREATE TABLE promocion (
    id_promocion INT PRIMARY KEY,
    nombre_mes VARCHAR(25) NOT NULL,
    fecha_comienzo DATE,
    id_campus INT,
    id_vertical INT,
    FOREIGN KEY (id_campus) REFERENCES campus(id_campus),
    FOREIGN KEY (id_vertical) REFERENCES vertical(id_vertical)
);
""")
```

```python
cur.execute("""
CREATE TABLE alumnos (
    id_alumno INT PRIMARY KEY,
    nombre VARCHAR(150) NOT NULL,
    email VARCHAR(150) NOT NULL,
    id_promocion INT,
    FOREIGN KEY (id_promocion) REFERENCES promocion(id_promocion)
);
""")
```

```python
cur.execute("""
CREATE TABLE profesores (
    id_profesor INT PRIMARY KEY,
    nombre_profesor VARCHAR(50) NOT NULL,
    rol VARCHAR(50) NOT NULL,
    modalidad VARCHAR(50) NOT NULL,
    id_promocion INT,
    FOREIGN KEY (id_promocion) REFERENCES promocion(id_promocion)
);
""")
```

```python
cur.execute("""
CREATE TABLE proyectos (
    id_proyecto INT PRIMARY KEY,
    nombre_proyecto VARCHAR(100) NOT NULL,
    id_vertical INT,
    FOREIGN KEY (id_vertical) REFERENCES vertical(id_vertical)
);
""")
```

```python
cur.execute("""
CREATE TABLE notas (
    id_alumno INT,
    id_proyecto INT,
    nota VARCHAR(50) NOT NULL,
    PRIMARY KEY (id_alumno, id_proyecto),
    FOREIGN KEY (id_alumno) REFERENCES alumnos(id_alumno),
    FOREIGN KEY (id_proyecto) REFERENCES proyectos(id_proyecto)
);
""")
```

```python
CREATE TABLE campus (
    id_campus SERIAL PRIMARY KEY,
    nombre_campus VARCHAR(25) NOT NULL
);

CREATE TABLE vertical (
    id_vertical SERIAL PRIMARY KEY,
    nombre_vertical VARCHAR(25) NOT NULL
);

CREATE TABLE proyectos (
    id_proyecto SERIAL PRIMARY KEY,
    nombre_proyecto VARCHAR(100) NOT NULL,
    id_vertical INT,
    CONSTRAINT fk_vertical_proy FOREIGN KEY (id_vertical) REFERENCES vertical(id_vertical)
);

CREATE TABLE promocion (
    id_promocion SERIAL PRIMARY KEY,
    nombre_mes VARCHAR(25) NOT NULL,
    fecha_comienzo DATE,
    id_campus INT,
    id_vertical INT,
    CONSTRAINT fk_campus_prom FOREIGN KEY (id_campus) REFERENCES campus(id_campus),
    CONSTRAINT fk_vertical_prom FOREIGN KEY (id_vertical) REFERENCES vertical(id_vertical)
);

CREATE TABLE alumnos (
    id_alumno SERIAL PRIMARY KEY,
    nombre VARCHAR(150) NOT NULL,
    email VARCHAR(150) NOT NULL,
    id_promocion INT,
    CONSTRAINT fk_promocion_alum FOREIGN KEY (id_promocion) REFERENCES promocion(id_promocion)
);

CREATE TABLE profesores (
    id_profesor SERIAL PRIMARY KEY,
    nombre_profesor VARCHAR(50) NOT NULL,
    rol VARCHAR(50) NOT NULL,
    modalidad VARCHAR(50) NOT NULL,
    id_promocion INT,
    CONSTRAINT fk_promocion_prof FOREIGN KEY (id_promocion) REFERENCES promocion(id_promocion)
);

CREATE TABLE notas (
    id_alumno INT,
    id_proyecto INT,
    nota VARCHAR(50) NOT NULL,
    PRIMARY KEY (id_alumno, id_proyecto),
    CONSTRAINT fk_alumno_nota FOREIGN KEY (id_alumno) REFERENCES alumnos(id_alumno),
    CONSTRAINT fk_proyecto_nota FOREIGN KEY (id_proyecto) REFERENCES proyectos(id_proyecto)
);
```
