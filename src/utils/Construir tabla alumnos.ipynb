```python
def generar_insert(df, tabla):
    valores = []
    for _, row in df.iterrows():
        valor = f"('{row['nombre']}', '{row['email']}', {row['id_promocion']})"
        valores.append(valor)
    
    query = f"INSERT INTO {tabla} (nombre, email, id_promocion) VALUES\n"
    query += ",\n".join(valores) + ";"
    
    return query

print(generar_insert(df_alumnos2, "Alumnos"))
```
